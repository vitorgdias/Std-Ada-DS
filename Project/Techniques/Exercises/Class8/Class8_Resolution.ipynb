```python
# Contar a quantidade de valores onde AcceptedCmp3 e AcceptedCmp4 são iguais a 1
count_cmp1 = df_full['AcceptedCmp1'].value_counts().get(1, 0)
count_cmp2 = df_full['AcceptedCmp2'].value_counts().get(1, 0)
count_cmp3 = df_full['AcceptedCmp3'].value_counts().get(1, 0)
count_cmp4 = df_full['AcceptedCmp4'].value_counts().get(1, 0)
count_cmp5 = df_full['AcceptedCmp5'].value_counts().get(1, 0)

# Criar um DataFrame para armazenar os resultados
df_counts = pd.DataFrame({
    'Campaign': ['AcceptedCmp1', 'AcceptedCmp2','AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5'],
    'Count': [count_cmp1, count_cmp2, count_cmp3, count_cmp4, count_cmp5]
})

plt.plot(df_counts['Campaign'], df_counts['Count'], marker='o')
plt.xlabel('Campaign')
plt.ylabel('Count')
plt.title('Quantidade de itens AcceptedCmp3 e AcceptedCmp4 = 1')
plt.grid(True)

# Mostrar o gráfico
plt.show()
```

```python
totais = df_full[['NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].sum()
df_totais = pd.DataFrame({
    'Categoria': ['Web Purchases', 'Catalog Purchases', 'Store Purchases'],
    'Total': totais.values
})

plt.bar(df_totais['Categoria'], df_totais['Total'], color=['blue', 'green', 'orange'])
plt.xlabel('Tipo de Compra')
plt.ylabel('Total de Compras')
plt.title('Total de Compras por Tipo')
plt.grid(axis='y')

plt.show()
```

```python
produtos = df_full[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum()

df_produtos_totais = pd.DataFrame({
    'Produto': ['Vinhos', 'Frutas', 'Carnes', 'Peixes', 'Doces', 'Produtos Gold'],
    'Total': produtos.values
})

# Criar gráfico de barras
plt.bar(df_produtos_totais['Produto'], df_produtos_totais['Total'], color=['brown', 'green', 'red', 'blue', 'pink', 'gold'])
plt.xlabel('Produto')
plt.ylabel('Total de Vendas')
plt.title('Total de Vendas por Produto')
plt.grid(axis='y')

# Mostrar o gráfico
plt.show()
```

```python
df_full['Dt_Customer'] = pd.to_datetime(df_full['Dt_Customer'])

df_full['YearMonth'] = df_full['Dt_Customer'].dt.to_period('M')

# Agrupar os dados por 'YearMonth' e calcular a média da Recency
df_grouped = df_full.groupby('YearMonth')['Recency'].mean().reset_index()

# Converter 'YearMonth' de volta para datetime para o gráfico
df_grouped['YearMonth'] = df_grouped['YearMonth'].dt.to_timestamp()

# Criar gráfico de linhas
plt.figure(figsize=(12, 6))
plt.plot(df_grouped['YearMonth'], df_grouped['Recency'], marker='o', linestyle='-', color='b')
plt.xlabel('Data de Cadastro')
plt.ylabel('Número Médio de Compras nos Últimos 2 Anos (Recency)')
plt.title('Número Médio de Compras nos Últimos 2 Anos em Função da Data de Cadastro (Agrupado por Mês)')
plt.grid(True)

# Mostrar o gráfico
plt.show()
```

```python
Kidhome = df_full['Kidhome'].value_counts().get(1, 0)
Teenhome = df_full['Teenhome'].value_counts().get(1, 0)

df_counts = pd.DataFrame({
    'KidsAndTeens': ['Kidhome', 'Teenhome'],
    'Count': [Kidhome, Teenhome]
})

plt.plot(df_counts['KidsAndTeens'], df_counts['Count'], marker='o')
plt.xlabel('KidsAndTeens')
plt.ylabel('Count')
plt.title('Quantidade de Kids e Teens = 1')
plt.grid(True)

plt.show()
```

```python
campaign_columns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']
df_full_campaign = df_full[campaign_columns].mean()
df_full_campaign.plot(kind='bar')
plt.title('Taxa de Adesão para Cada Campanha')
plt.xlabel('Campanha')
plt.ylabel('Taxa de Adesão')
plt.show()

# Pergunta 3: Perfil dos usuários que mais gastaram
df_full['TotalSpent'] = df_full[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)
df_full_grouped = df_full.groupby('Education')['TotalSpent'].mean()
df_full_grouped.plot(kind='bar')
plt.title('Gasto Médio por Nível de Educação')
plt.xlabel('Educação')
plt.ylabel('Gasto Médio')
plt.show()

# Pergunta 6: Quem mais ganha é quem mais gasta?
df_full.plot(kind='scatter', x='Income', y='TotalSpent')
plt.title('Renda vs Gasto Total')
plt.xlabel('Renda')
plt.ylabel('Gasto Total')
plt.show()
```
