```python
import pandas as pd
import numpy as np
```

```python
#1 Utilizando o DataFrame importado anteriormente (alunos3.csv) calcule a média das provas em uma nova coluna chamada (Media_provas)
df = pd.read_csv('./alunos3.csv', sep=';')
df[['Prova_1', 'Prova_2', 'Prova_3', 'Prova_4']] = df[['Prova_1', 'Prova_2', 'Prova_3', 'Prova_4']].replace(',', '.', regex=True)
df[['Prova_1', 'Prova_2', 'Prova_3', 'Prova_4']] = df[['Prova_1', 'Prova_2', 'Prova_3', 'Prova_4']].astype(float)

df['media_provas'] = round(df[['Prova_1', 'Prova_2', 'Prova_3', 'Prova_4']].mean(axis=1),1)
df
```

```python
# 2) Quem foram os alunos que obtiveram a maior e a menor média
#Maior média
df[df['media_provas'] == df['media_provas'].max()]
```

```python
#Menor média
df[df['media_provas'] == df['media_provas'].min()]
```

```python
# 3) Agora una este dataframe com o cadastro_alunos.xlsx
df_cadastro = pd.read_excel('./cadastro_alunos.xlsx')
df_merged = df.merge(df_cadastro)
df_merged
```

```python
# 4) Qual a média entre as Media_provas dentro do público feminino? e masculino?
notas_femininas = df_merged[df_merged['Sexo'] == 'Feminino']
media_fem = notas_femininas['media_provas'].mean(axis=0)

notas_masculinas = df_merged[df_merged['Sexo'] == 'Masculino']
media_masc = notas_masculinas['media_provas'].mean(axis=0)

print(f'Média das notas femininas: {media_fem.round(1)}')
print(f'Média das notas masculinas: {media_masc.round(1)}')
```

```python
# 5) Qual a média de idade das pessoas que obtiveram Media_provas maior ou igual a 7?
media_aprovados = df_merged[df_merged['media_provas'] >= 7]

media_idade = media_aprovados['Idade'].mean()
print(f'Média das idades dos aprovados: {media_idade}')
```

```python
# 6) Qual das cidades possui o maior média de Media_provas? E qual é este valor?
media_cidade = df_merged.groupby('Cidade')['media_provas'].mean()
cidade_maior_media = media_cidade.idxmax()
maior_media = media_cidade.max()


print(f'A cidade com a maior média é: {cidade_maior_media}, com média de: {maior_media.round(1)}')
```
