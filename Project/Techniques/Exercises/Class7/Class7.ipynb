## 2 - Juntar os dataframes de investimento1 e investimento2 em um só e ver a quantidade de linhas

```python
full_inv = pd.concat([df_inv1, df_inv2])
full_inv.head()
```

## 3 - Crie um novo dataframe eliminando as linhas duplicadas do item anterior

```python
inv_refactored = full_inv.drop_duplicates()
```

```python
print(f"Após eliminar duplicatas, o dataframe tem {inv_refactored.shape[0]} linhas.")
```

## 4 - Junte o dataframe de investimento com o de cadastro, criando um novo dataframe
Dica: use a coluna id como chave para juntar os dois dataframes

```python
df_merged = pd.merge(inv_refactored, df_cadastro, on='id')
```

```python
df_merged.head()
```

## 5 - Ordene o dataframe pelo id, de forma crescente e redefina o índice

```python
df_merged = df_merged.sort_values(by='id').reset_index(drop=True)
df_merged.head()
```

## 6 - Faça uma análise dos dados faltantes e trate-os
Mostre todas as colunas com dados faltantes, sua quantidade/porcentagem e como tratar (preencher) cada uma delas explicando sua estratégia

```python
missing_data = df_merged.isnull().sum()
total_cells = df_merged.shape[0] * df_merged.shape[1]
missing_percentage = (missing_data / df_merged.shape[0]) * 100

# Mostra as colunas com dados faltantes e suas respectivas porcentagens
missing_info = missing_data[missing_data > 0]
print(missing_info)

# Tratamento dos dados
for col in missing_info.index:
    if df_merged[col].dtype == 'float64' or df_merged[col].dtype == 'int64':
        df_merged[col].fillna(df_merged[col].mean())
    else:
        df_merged[col].fillna('Desconhecido', inplace=True)
```

```python
df_merged.head()
```

## 7 - Manipule os dados a seguir:
Criar coluna "Total" somando o valor de todos os ativos por investidor

Criar coluna "Media" com a média das notas 1 e 2

Criar coluna "Dominio" com apenas o domínio do email (dica: separar por @)

Mudar o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter $ e tipo numérica)

Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar a função map)

Transformar os dados da coluna Media em inteiro

Criar coluna "NPS" onde o investidor é classificado como promotor se a média das notas é maior ou igual a 80, como neutro se a média estiver entre 50 e 80, e como detrator se for abaixo de 50

```python
# Criando a coluna 'Total'
df_merged['Total'] = df_merged[['Ativo01', 'Ativo02', 'Ativo03', 'Ativo04', 'Ativo05']].sum(axis=1)

# Criando a coluna 'Media'
df_merged['Media'] = df_merged[['Nota1', 'Nota2']].mean(axis=1)

# Criando a coluna 'Dominio'
df_merged['Dominio'] = df_merged['Email'].str.split('@').str[1]

# Renomeando a coluna 'Salário'e tratando os dados
df_merged['Renda'] = df_merged['Salario'].replace({'\$': '', ',': ''}, regex=True).astype(float)

# Tratando a coluna 'Sexo'
df_merged['Sexo'] = df_merged['Sexo'].map({'Masc': 'Masculino', 'Fem': 'Feminino'})

# Preencherdo os NaN com 0 e convertendo a coluna 'Media' para inteiro
df_merged['Media'] = df_merged['Media'].fillna(0).astype(int)

# Criando a coluna 'NPS' e classificando como promotor, neutro ou detrator com base na 'Media'
df_merged['NPS'] = df_merged['Media'].apply(lambda x: 'Promotor' if x >= 80 else 'Neutro' if 50 <= x < 80 else 'Detrator')
```

## 8 - Calcule os seguintes dados
Calcule o total investido em cada coluna Ativo (tente usar o apply)

Qual o menor valor investido em cada ativo?

Mostre a média e desvio padrão de cada ativo

Qual o nome e a idade do investidor mais velho?

Qual a mediana da coluna Média?

```python
# Total investido em cada ativo
total_investido = df_merged[['Ativo01', 'Ativo02', 'Ativo03', 'Ativo04', 'Ativo05']].apply(lambda x: x).sum()

# Menor valor investido em cada ativo
menor_investido = df_merged[['Ativo01', 'Ativo02', 'Ativo03', 'Ativo04', 'Ativo05']].min()

# Média e desvio padrão de cada ativo
media_ativos = df_merged[['Ativo01', 'Ativo02', 'Ativo03', 'Ativo04', 'Ativo05']].mean()
desvio_padrao_ativos = df_merged[['Ativo01', 'Ativo02', 'Ativo03', 'Ativo04', 'Ativo05']].std()

# Nome e idade do investidor mais velho
investidor_mais_velho = df_merged.loc[df_merged['Idade'].idxmax(), ['Nome', 'Idade']]

# Mediana da coluna Média
mediana_media = df_merged['Media'].median()
```

```python
# Calcule o total investido em cada coluna Ativo (tente usar o apply)
print(f'O total investido em cada ativo é: \n{total_investido}')
```

```python
# Qual o menor valor investido em cada ativo?
print(f'O menor valor investido de cada ativo é: \n{menor_investido}')
```

```python
# Mostre a média e desvio padrão de cada ativo
print(f'A média de cada ativo é: \n{media_ativos}')
print(30*'-')
print(f'O desvio padrão de cada ativo é: \n{desvio_padrao_ativos}')
```

```python
# Qual o nome e a idade do investidor mais velho?
print(f'O investidor mais velho é: \n{investidor_mais_velho}')
```

```python
# Qual a mediana da coluna Média?
print(f'A mediana da coluna média é: \n{mediana_media}')
```

## 9 - Responder as perguntas abaixo
Quantos investidores são agressivos, moderados e conservadores?

Mostre os 10 primeiros ids dos investidores de perfil Agressivo e Moderado.

Quantos investidores são do sexo feminino?

Quantos investidores investiram mais de 16000 no Ativo04?

Quantos investidores investiram mais de 16000 no Ativo04 e menos de 5000 no Ativo05?

Quais investidores do sexo feminino tem mais de 25 anos e seu investimento total foi menor que 10000?

Quantos investidores tem em cada estado (dica: use groupby)

Qual é a soma do total investido por cada estado?

Qual é a média do total investido por promotores, detratores e neutros? E como se dividem nos estados

Crie 3 colunas "dummies" do perfil dos investidores

Crie uma tabela com total de investimentos de promotores, detratores e neutros em função do perfil. (dica: use pivot_table)

```python
# Quantidade de investidores por perfil
perfis = df_merged['Perfil'].value_counts()

# IDs dos primeiros 10 investidores agressivos e moderados
ids_agressivos = df_merged[df_merged['Perfil'] == 'Agressivo'].head(10)['id']
ids_moderados = df_merged[df_merged['Perfil'] == 'Moderado'].head(10)['id']

# Quantidade de investidores do sexo feminino
femininos = df_merged[df_merged['Sexo'] == 'Feminino'].shape[0]

# Investidores com mais de 16000 no Ativo04
inv_ativo04 = df_merged[df_merged['Ativo04'] > 16000].shape[0]

# Investidores com mais de 16000 no Ativo04 e menos de 5000 no Ativo05
inv_ativo04_05 = df_merged[(df_merged['Ativo04'] > 16000) & (df_merged['Ativo05'] < 5000)].shape[0]

# Investidores femininos com mais de 25 anos e total menor que 10000
inv_fem_25 = df_merged[(df_merged['Sexo'] == 'Feminino') & (df_merged['Idade'] > 25) & (df_merged['Total'] < 10000)]

# Quantidade de investidores por estado
inv_estado = df_merged.groupby('Estado').size()

# Soma total investido por estado
total_estado = df_merged.groupby('Estado')['Total'].sum()

# Média total investido por promotores, detratores e neutros
media_nps = df_merged.groupby('NPS')['Total'].mean()

# Divisão por estados
divisao_nps_estado = df_merged.groupby(['NPS', 'Estado'])['Total'].sum()

# Criação das colunas dummies
dummies_perfil = pd.get_dummies(df_merged['Perfil'])

# Tabela pivot
pivot_nps_perfil = pd.pivot_table(df_merged, values='Total', index='NPS', columns='Perfil', aggfunc='sum')
```
